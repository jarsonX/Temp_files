**Import danych**

2. Create python code which will download data from csv and save it in datarfame. Use columns Age, City,Country, Bib, M/F. Make Bib column index. 

```python
columns = ["Age", 
           "City", 
           "Country", 
           "Bib", 
           "M/F"]

df = pd.read_csv("raw.csv", index_col="Bib", usecols=columns)
df.head()
```

**Analiza**

3. Create gropus based on Country and City column

```python
df_countries_cities = df.groupby(["Country", "City"], as_index=False)
```

4. Calculate mean age for each groups.

```python
df_countries_cities["Age"].mean()

# Kolumna "City" wymaga uspójnienia (np. "Ciudad Autonoma De Buenos Aires" vs
# "Ciudad De Buenos Aires"). Nie przychodzi mi do główy żaden szybki sposób na 
# naprawienie tego, więc zakładam, że celem zadania było samo grupowanie 
# i policzenie średniej.
```

5. Display results from group JPN, Tokyo

```python
df_JPN = df_countries_cities.get_group(("JPN", "Tokyo"))
df_JPN
```

```python
# Średnia dla grupy JPN, Tokyo
df_JPN["Age"].mean()
```

6. Using dataframe from step **5**, make selection and display results only for values diffrent than F in M/F column

```python
df_JPN[df_JPN["M/F"] != "F"]
```

7. Using dataframe from step **6**, display results for Age between 50 and 70. For this step you may write only one line of code (one string of characters).

```python
df_to_save = df_JPN[(df_JPN["Age"] > 50) & (df_JPN["Age"] < 70)]
df_to_save
```


8. Save results from step **7** in csv file. There should not be an index in newly created csv.

```python
df_to_save.to_csv("JPN_Tokyo_Age_50_70")
```
